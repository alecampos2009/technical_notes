```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)
rc('font', family='serif')

params = {'font.size' : 22.0,
          'legend.fontsize': 22.0,
          'axes.labelsize': 25.0,
          'axes.titlesize': 40.0,
          'xtick.labelsize': 32.0,
          'ytick.labelsize': 32.0,
          'lines.linewidth': 4,
          'figure.figsize': (8, 6),}

plt.rcParams.update(params)
```

```python
# Debye length

num_pts = 101
L_lam_1 = 40
L_lam_2 = 20
L_lam_3 = 10

x_L = np.linspace(-0.5, 0.5, num_pts)

phi_1 = -0.5 * np.sinh(L_lam_1 * x_L) / np.sinh(L_lam_1 * 0.5)
elec_1 = 0.5 * np.cosh(L_lam_1 * x_L) / np.sinh(L_lam_1 * 0.5)

phi_2 = -0.5 * np.sinh(L_lam_2 * x_L) / np.sinh(L_lam_2 * 0.5)
elec_2 = 0.5 * np.cosh(L_lam_2 * x_L) / np.sinh(L_lam_2 * 0.5)

phi_3 = -0.5 * np.sinh(L_lam_3 * x_L) / np.sinh(L_lam_3 * 0.5)
elec_3 = 0.5 * np.cosh(L_lam_3 * x_L) / np.sinh(L_lam_3 * 0.5)

fig, ax = plt.subplots()

ax.plot(x_L, phi_1, label = f'$\lambda_D/L = {1.0 / L_lam_1}$')
ax.plot(x_L, phi_2, label = f'$\lambda_D/L = {1.0 / L_lam_2}$')
ax.plot(x_L, phi_3, label = f'$\lambda_D/L = {1.0 / L_lam_3}$')

ax.set_xlabel(r'$x/L$')
ax.set_ylabel(r'$\phi/V$')
ax.tick_params(axis='both', top=True, right=True, direction='in')
ax.legend()

fig.tight_layout()
fig.savefig('debye_phi.pdf')

fig, ax = plt.subplots()

ax.plot(x_L, elec_1)
ax.plot(x_L, elec_2)
ax.plot(x_L, elec_3)

ax.set_xlabel(r'$x/L$')
ax.set_ylabel(r'$E\lambda_D/V$')
ax.tick_params(axis='both', top=True, right=True, direction='in')

fig.tight_layout()
fig.savefig('debye_elec.pdf')
```

```python
# Plasma frequency

num_pts = 101
L_alpha_1 = 40
L_alpha_2 = 20
L_alpha_3 = 10

x_L = np.linspace(-0.5, 0.5, num_pts)

phi_1 = -0.5 * np.sin(L_alpha_1 * x_L) / np.sin(L_alpha_1 * 0.5)
elec_1 = 0.5 * np.cos(L_alpha_1 * x_L) / np.sin(L_alpha_1 * 0.5)

phi_2 = -0.5 * np.sin(L_alpha_2 * x_L) / np.sin(L_alpha_2 * 0.5)
elec_2 = 0.5 * np.cos(L_alpha_2 * x_L) / np.sin(L_alpha_2 * 0.5)

phi_3 = -0.5 * np.sin(L_alpha_3 * x_L) / np.sin(L_alpha_3 * 0.5)
elec_3 = 0.5 * np.cos(L_alpha_3 * x_L) / np.sin(L_alpha_3 * 0.5)

fig, ax = plt.subplots()

ax.plot(x_L, phi_1, label = fr'$\alpha/L = {1.0 / L_alpha_1}$')
ax.plot(x_L, phi_2, label = fr'$\alpha/L = {1.0 / L_alpha_2}$')
ax.plot(x_L, phi_3, label = fr'$\alpha/L = {1.0 / L_alpha_3}$')

ax.set_xlabel(r'$x/L$')
ax.set_ylabel(r'$\phi/V$')
ax.tick_params(axis='both', top=True, right=True, direction='in')
ax.legend()

fig.tight_layout()
fig.savefig('pf_phi.pdf')

fig, ax = plt.subplots()

ax.plot(x_L, elec_1)
ax.plot(x_L, elec_2)
ax.plot(x_L, elec_3)

ax.set_xlabel(r'$x/L$')
ax.set_ylabel(r'$E\alpha/V$')
ax.tick_params(axis='both', top=True, right=True, direction='in')

fig.tight_layout()
fig.savefig('pf_elec.pdf')
```
